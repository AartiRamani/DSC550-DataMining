```python
#Create a new column in the data frame that calculates the percentage of global sales from North America. 
#Display the first five rows of the new data frame.
 
video_games_df['NA_Sales_Percentage'] = round(video_games_df.NA_Sales/video_games_df.Global_Sales*100,2)
video_games_df.head(5)
```

```python
#Find the number NaN entries (missing data values) in each column.
video_games_df.isnull().values.any()
video_games_df.isnull().sum()
```

```python
#Try to calculate the median user score of all the video games. 
#You will likely run into an error because some of the user score entries are a non-numerical string 
#that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. 
video_games_df[pd.to_numeric(video_games_df['User_Score'], errors='coerce').isnull()]['User_Score'].unique()
```

```python
video_games_df.User_Score = video_games_df.User_Score.replace('tbd',np.nan)
median = video_games_df.User_Score.median()
print('Median after replacing string with Nan : ',median)

#Then, replace all NaN entries in the user score column with the median value.
video_games_df.User_Score = video_games_df.User_Score.replace(np.nan,median) 
```
